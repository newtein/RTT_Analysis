```python
import xml.etree.ElementTree as et
import pickle
import collections
from collections import OrderedDict
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import numpy as np
import matplotlib.ticker as ticker
```

```python
picklename=[1,2,3,4,5,6]
filename=["00-04","04-08","08-12","12-16","16-20","20-00"]
```

```python
data={}
for fname,alias in zip(filename,picklename):
    data=OrderedDict()
    tree=et.parse("Data/"+fname+".xml")
    root=tree.findall("host")

    for host in root:
        os,acc=None,None
        ipv4=host.find(".//address[@addrtype='ipv4']").get("addr")
        osnode=host.find(".os/osmatch")
        if osnode!=None:
            os=osnode.get("name")
            acc=osnode.get("accuracy")
        print(ipv4,os,acc)
        if ipv4 not in data:
            data[ipv4]=[]
        data[ipv4].append(os)
group={}
for ipv4 in data:
    print(ipv4, str(list(set(data[ipv4]))[0]).split("or")[0])
    key=str(list(set(data[ipv4]))[0]).split("or")[0]
    data[ipv4]=str(list(set(data[ipv4]))[0]).split("or")[0]
    if key not in group:
        group[key]=0
    group[key]=group[key]+1
print(group)

plt.close()
plt.clf()
fig, ax = plt.subplots(figsize=(9,6))
#plt.rcParams['axes.formatter.useoffset']=False


xx=[i for i in range(len(data))]

x=[i for i in data]
y=[i]
ax.bar(xx,y, label="Uptime (since last boot)")
#ax.scatter(xx,[i+500000 for i in y],s=z,label="Reason TTL (diameter)")

plt.xticks(xx,x,rotation=30)

plt.title("Uptime and TTL of Hosts")
plt.xlabel("Ipv4 address of host")
plt.ylabel("Time (in secs)")
plt.legend(loc="upper right")
plt.show()
```
