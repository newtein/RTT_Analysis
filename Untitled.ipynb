```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score


# In[2]:


df = pd.read_csv('ipid_prediction.csv',low_memory=False)
```

```python

df=df[df[" ipidclass"]!=" None"]
```

```python
df[' ipidclass'] = df[' ipidclass'].map({' Incremental': 0, ' All zeros': 1,' Busy server or unknown class':2, ' Incrementing by 2':3})
```

```python
X = df.drop(' ipidclass', axis=1)
X = X.convert_objects(convert_numeric=True)
X = X._get_numeric_data()
y = df[' ipidclass']
X=(X-X.mean())/X.std()
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1, test_size = .35)

model = tree.DecisionTreeClassifier()
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
accuracy_score(y_test, y_predict)
```

```python
from sklearn.datasets import load_iris
from sklearn import tree
import collections
from sklearn.externals.six import StringIO
from IPython.display import Image
from sklearn.tree import export_graphviz
import pydotplus
from sklearn.datasets import load_iris
from sklearn import tree
import collections
dot_data = tree.export_graphviz(model, feature_names=['rtt','srtt','rttvar','to','scantime'], out_file=None, filled=True, rounded=True)
graph = pydotplus.graph_from_dot_data(dot_data)
colors = ('turquoise', 'orange')
edges = collections.defaultdict(list)
for edge in graph.get_edge_list():
    edges[edge.get_source()].append(int(edge.get_destination()))
for edge in edges:
    edges[edge].sort()
    for i in range(2):
        dest = graph.get_node(str(edges[edge][i]))[0]
        dest.set_fillcolor(colors[i])
graph.write_png('tree.png')
```
